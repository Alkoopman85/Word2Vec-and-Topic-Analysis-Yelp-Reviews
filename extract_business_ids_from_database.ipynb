```python
from database import YelpDataBase

def show_important_categories(limit:int=20):
    ydb = YelpDataBase()
    
    statement = f"""SELECT category.name, COUNT(reviews.id) AS num_reviews, COUNT(DISTINCT business.id) AS num_business
                FROM reviews JOIN business ON reviews.business_id  == business.id
                JOIN category_business ON business.id == category_business.business_id
                JOIN category ON category_business.category_id == category.id
                GROUP BY category.name ORDER BY num_reviews DESC LIMIT {limit}"""
    frame = ydb.execute_sql_statement(statement, as_frame=True)
    return frame

def get_business_ids_category(categories):
    ydb = YelpDataBase()

    cat_statement = ' OR '.join([f"category.name == '{cat}'" for cat in categories])
    
    statement = f"""SELECT business.business_id_str FROM business
                JOIN category_business ON business.id == category_business.business_id
                JOIN category ON category_business.category_id == category.id
                WHERE {cat_statement}"""
    frame = ydb.execute_sql_statement(statement, as_frame=False)
    return frame
```

```python
important_cats = show_important_categories(100)
important_cats
```
